```python
import pandas as pd
import joblib

loaded_model = joblib.load("anemia_model.pkl")
loaded_le = joblib.load("label_encoder.pkl")
loaded_columns = joblib.load("model_columns.pkl")

sample_data = {
    "Age in 5-year groups": "20-24",
    "Type of place of residence": "Urban",
    "Highest educational level": "Secondary",
    "Wealth index combined": "Middle",
    "Births in last five years": 1,
    "Age of respondent at 1st birth": 20,
    "Have mosquito bed net for sleeping (from household questionnaire)": "Yes",
    "Smokes cigarettes": "No",
    "Current marital status": "Married",
    "Currently residing with husband/partner": "Yes"
}

input_df = pd.DataFrame([sample_data])
input_encoded = pd.get_dummies(input_df)

for col in loaded_columns:
    if col not in input_encoded.columns:
        input_encoded[col] = 0

input_encoded = input_encoded[loaded_columns]

pred = loaded_model.predict(input_encoded)
label = loaded_le.inverse_transform(pred)

print(pred[0])
print(label[0])
```
